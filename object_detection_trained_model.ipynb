```python
import os
from ultralytics import YOLO
from IPython.display import display, Image
from IPython import display
display.clear_output()

# set up environment
os.environ["DATASET_DIRECTORY"] = "/content/datasets"

!pip install roboflow

from roboflow import Roboflow
rf = Roboflow(api_key="rE192HmD2H2lJJ4SZOxD")
project = rf.workspace("thesis-q0sbh").project("supershop_products_dataset")
version = project.version(4)
dataset = version.download("yolov8")

# Initialize YOLO object detection model
model = "/content/runs/detect/train/weights/best.pt"

# Evaluate on validation set
results = model.evaluate(data=f"{dataset.location}/content/datasets/supershop_products_dataset-4/data.yaml")
print(results)

# Perform prediction on test set
predictions = model.predict(source=f"{dataset.location}/test/images")

# You can further process predictions if needed, e.g., draw bounding boxes on images
# predictions.show()

# If you want to calculate additional metrics like precision, recall, and F1 score manually,
# you can utilize the results dictionary obtained from model.evaluate()

# Confusion matrix
confusion_matrix = results["confusion_matrix"]
print("Confusion Matrix:")
print(confusion_matrix)

# Precision, Recall, F1 score
precision = results["precision"]
recall = results["recall"]
f1_score = results["f1"]
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```
